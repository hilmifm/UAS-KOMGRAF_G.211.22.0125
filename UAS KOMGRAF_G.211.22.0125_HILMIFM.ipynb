1. A. Library Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data curah hujan
tahun = list(range(1981, 2011))
curah_hujan = [485, 482, 422, 305, 420, 415, 289, 420, 367, 260,
               340, 188, 140, 122, 125, 122, 40, 50, 40, 70,
               60, 20, 225, 400, 235, 420, 490, 400, 410, 430]


# Membuat dataframe
data = {'Tahun': tahun, 'Curah Hujan': curah_hujan}
df = pd.DataFrame(data)

# Membuat plot menggunakan seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Tahun', y='Curah Hujan')

# Menambahkan judul dan label sumbu
plt.title('Fluktuasi Curah Hujan (1981-2010)', fontsize=16)
plt.xlabel('Tahun')
plt.ylabel('Curah Hujan (mm)')

# Menampilkan plot
plt.show()
```

1. B. Matplotlib

```python
import matplotlib.pyplot as plt

# Data curah hujan
tahun = list(range(1981, 2011))
curah_hujan = [485, 482, 422, 305, 420, 415, 289, 420, 367, 260,
               340, 188, 140, 122, 125, 122, 40, 50, 40, 70,
               60, 20, 225, 400, 235, 420, 490, 400, 410, 430]

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(tahun, curah_hujan, marker='o', linestyle='-', color='b')

# Menambahkan judul dan label sumbu
plt.title('Fluktuasi Curah Hujan (1981-2010)', fontsize=16)
plt.xlabel('Tahun')
plt.ylabel('Curah Hujan (mm)')

# Menampilkan grid
plt.grid(True)

# Menampilkan plot
plt.show()
```

2. A. Matplotlib


```python
import matplotlib.pyplot as plt

# Data suhu udara observasi (misalkan dalam Celsius)
observasi = [24, 25, 23.5, 24, 24, 24.2, 24, 22.8, 23]

# Data suhu udara WRF non Asimilasi (misalkan dalam Celsius)
wrf_non_asimilasi = [23.5, 25.5, 25.2, 25.5, 24.8, 25, 24.6, 24.2, 25.8]

# Data suhu udara WRF Asimilasi (misalkan dalam Celsius)
wrf_asimilasi = [23.2, 25.3, 25.5, 24.8, 22.5, 22.2, 23, 22, 25.7]

# Membuat plot
plt.figure(figsize=(10, 6))
plt.plot(range(1, 10), observasi, marker='o', linestyle='-', color='b', label='Observasi')
plt.plot(range(1, 10), wrf_non_asimilasi, marker='o', linestyle='-', color='r', label='WRF non Asimilasi')
plt.plot(range(1, 10), wrf_asimilasi, marker='o', linestyle='-', color='g', label='WRF Asimilasi')

# Menambahkan judul dan label sumbu
plt.title('Perbandingan Suhu Udara (24-25 Juli 2013)', fontsize=16)
plt.xlabel('Jam')
plt.ylabel('Suhu Udara (Celsius)')

# Menampilkan plot
plt.show()
```

2. B. Library Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data suhu udara observasi (misalkan dalam Celsius)
observasi = [24, 25, 23.5, 24, 24, 24.2, 24, 22.8, 23]

# Data suhu udara WRF non Asimilasi (misalkan dalam Celsius)
wrf_non_asimilasi = [23.5, 25.5, 25.2, 25.5, 24.8, 25, 24.6, 24.2, 25.8]

# Data suhu udara WRF Asimilasi (misalkan dalam Celsius)
wrf_asimilasi = [23.2, 25.3, 25.5, 24.8, 22.5, 22.2, 23, 22, 25.7]

# Membuat dataframe
data = {'Jam': list(range(1, 10)),
        'Observasi': observasi,
        'WRF non Asimilasi': wrf_non_asimilasi,
        'WRF Asimilasi': wrf_asimilasi}
df = pd.DataFrame(data)

# Membuat plot menggunakan seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Jam', y='Observasi', marker='o', linestyle='-', color='b', label='Observasi')
sns.lineplot(data=df, x='Jam', y='WRF non Asimilasi', marker='o', linestyle='-', color='r', label='WRF non Asimilasi')
sns.lineplot(data=df, x='Jam', y='WRF Asimilasi', marker='o', linestyle='-', color='g', label='WRF Asimilasi')

# Menambahkan judul dan label sumbu
plt.title('Perbandingan Suhu Udara (24-25 Juli 2013)', fontsize=16)
plt.xlabel('Jam')
plt.ylabel('Suhu Udara (Celsius)')

# Menampilkan plot
plt.show()
```
